```python
image = "image_4_20250409220505_a98b7f"
folder = "samples_multi2"
img_path = f"output/{folder}/images/{image}.png"
bbox_path = f"output/{folder}/labels/{image}_bbox.txt"
```

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw

img = Image.open(img_path)


with open(bbox_path, 'r') as f:
    bbox_data = f.readline().strip().split()
   
    x1 = float(bbox_data[0])  
    y1 = float(bbox_data[1]) 
    width = float(bbox_data[2])
    height = float(bbox_data[3])
    
    x2 = x1 + width
    y2 = y1 + height

img_with_bbox = img.copy()
draw = ImageDraw.Draw(img_with_bbox)

draw.rectangle([x1, y1, x2, y2], outline="red", width=2)

plt.figure(figsize=(10, 8))
plt.imshow(img_with_bbox)
plt.title("Image with Bounding Box")
plt.axis('off')
plt.show()

print(f"Bounding Box Information:")
print(f"Top-left coordinates (x, y): ({x1:.2f}, {y1:.2f})")
print(f"Width: {width:.2f}, Height: {height:.2f}")
```
